```python
import os
import pdfplumber
import requests
from bs4 import BeautifulSoup
from transformers import pipeline
import streamlit as st

# Define the PDF extraction function
def extract_text_from_pdf(file_path):
    text = ""
    with pdfplumber.open(file_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    return text

# Define the text file extraction function
def extract_text_from_txt(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

# Define the function to extract text based on file extension
def extract_text_from_file(file_path):
    if file_path.endswith('.pdf'):
        return extract_text_from_pdf(file_path)
    elif file_path.endswith('.txt'):
        return extract_text_from_txt(file_path)
    else:
        return ""

# Define the function to read files from a directory
def read_files_from_directory(directory_path):
    file_texts = []
    for root, _, files in os.walk(directory_path):
        for file in files:
            file_path = os.path.join(root, file)
            text = extract_text_from_file(file_path)
            if text:
                file_texts.append(text)
    return file_texts

# Define the web data fetching function
def fetch_web_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup.get_text()

# Define the knowledge base class
class KnowledgeBase:
    def __init__(self):
        self.data = []

    def add_data(self, text):
        self.data.append(text)

    def search(self, query):
        results = [d for d in self.data if query.lower() in d.lower()]
        return results

# Initialize the Hugging Face pipeline for text generation
generator = pipeline('text-generation', model='gpt2')

def generate_response(prompt):
    responses = generator(prompt, max_length=150, num_return_sequences=1)
    return responses[0]['generated_text']

# Define the DocuMaster class
class DocuMaster:
    def __init__(self):
        self.kb = KnowledgeBase()

    def learn_from_directory(self, directory_path):
        texts = read_files_from_directory(directory_path)
        for text in texts:
            self.kb.add_data(text)

    def get_real_time_solution(self, query, web_url=None):
        kb_results = self.kb.search(query)
        web_data = fetch_web_data(web_url) if web_url else ""
        combined_data = " ".join(kb_results) + " " + web_data
        response = generate_response(combined_data + " " + query)
        return response

# Initialize DocuMaster
docu_master = DocuMaster()

# Define Streamlit app
st.title("DocuMaster AI Chat")

# Sidebar for learning from directory
st.sidebar.title("Learn from Directory")
directory_path = st.sidebar.text_input("Directory Path", "")
if st.sidebar.button("Learn"):
    if os.path.isdir(directory_path):
        docu_master.learn_from_directory(directory_path)
        st.sidebar.success(f"Learned from directory: {directory_path}")
    else:
        st.sidebar.error(f"Invalid directory: {directory_path}")

# Main chat interface
query = st.text_input("Ask something:", "")
web_url = st.text_input("Web URL (optional):", "")
if st.button("Get Answer"):
    if query:
        answer = docu_master.get_real_time_solution(query, web_url)
        st.write("Answer:", answer)
    else:
        st.write("Please enter a query.")
```
